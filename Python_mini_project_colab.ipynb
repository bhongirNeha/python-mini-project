## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyse and visualise data. The code cell below uses <strong>numpy</strong> to generate some random data, and uses <strong>matplotlib</strong> to visualise it. To edit the code, just click the cell and start editing.

You can import your own data into Colab notebooks from your Google Drive account, including from spreadsheets, as well as from GitHub and many other sources. To find out more about importing data, and how Colab can be used for data science, see the links below under <a href="#working-with-data">Working with data</a>.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

#PART 4: Visualization using Matplotlib

##Matplotlib is imported for plotting graphs.

```
import matplotlib.pyplot as plt
```

##CHART 1-LINE CHART

The dimensions of graph size is mentioned in the 1st line of the code. The second line describes about the index and the column name, 'x' is the marker given to print the points in the graph.Title,xlabel and ylabel are given.

```
plt.figure(figsize=(8,4))
plt.plot(df.index, df['sepal_length'], marker = 'x')
plt.title('Sepal Length Line Chart')
plt.xlabel('Index')
plt.ylabel('Sepal Length')
plt.show()
```

###INTERPRETATION
Line chart shows the trend.The above graph is depicting the trend of sepal length across all the species in the data.The length of the sepal is varing among the species.

##TASK 2 BAR CHART

Value counts() gives the total species counts.

```
species_count= df['species'].value_counts()
species_count
```

The dimensions,title,xlabel and ylabel are given.The second line code accesses the index and count of species in frequency table.

```
plt.figure(figsize=(8,4))
plt.bar(species_count.index,species_count)
plt.title(' Total Species Count')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

### INTERPRETATION
Bar chart is used for the categorical data. The graph shows total species count for each species.Here, the 3 species have same counts of around 50.

##TASK 3-SCATTER PLOT

Here the extra code called grid is given so that the grid is displayed in the graph so that the data points can be analysed accurately in the range it is falling in.

```
plt.figure(figsize=(8,4))
plt.scatter(df['sepal_length'], df['petal_length'])
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.grid(True)
plt.show()
```

###INTERPRETATION
The above Scatter plot shows the correlation between sepal length and petal length of all the species.One of the species is falling in lower range and the rest species are falling in the higher ranges.


##CHART 4- Histogram

Bins is extra code we give in Histogram from the rest of the common codes for the graphs.This code divides the data into 20 parts.

```
plt.figure(figsize=(8,4))
plt.hist(df['sepal_width'], bins=20)
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.show()
```

### INTERPRETATION

Histogram shows the distribution of the data. The above shows the distribution of sepal width across all the species in the data.Sepal width is varying among different species.

#PART 5: Visualization using plotly

##CHART 1-Interactive Scatter plot





In plotly instead of x & y labels the column names from the dataframe are assigned directly to x and y axis.The code show() displays the graph.

```
fig = px.scatter(
    df,
    x='sepal_length',
    y='petal_length',
    color='species',
    title='Interactive Scatter Plot'
)
fig.show()
```

###INTERPRETATION

Scatter plot gives the correlation between the variables. The spread of the data can be observed in the above plot were there are outliers and overlapping points present.Here,the relation between the sepal and petal length is seen.Setosa species have smaller petal and sepal length when compared to versicolor and virginica.Virginica shows larger petal and sepal length.

##CHART 2-BOX PLOT

In box plot, color is given to represent different species with different colors.

```
fig = px.box(
    df,
    x='species',
    y='petal_length',
    color='species',
    title='Box Plot of Petal length by Species'
)
fig.show()
```

###INTERPRETATION

Box plot is used to analyze the outliers.The graph shows the petal length across all the species.Outliers for versicolor and setosa species are observed.Rest of the data points are falling within the range.

## CHART 3-Histogram

```
fig = px.histogram(
    df,
    x='sepal_length',
    nbins=20,
    title='Distribution of Sepal Length'
)
fig.show()
```

###INTERPRETATION
Histogram shows the frequency distribution of the data.The graph shows the distribution of sepal length across all species.Here,the data seems to look like it is slightly going beyond the normal distribution in the initial points.

##CHART 4-PIE CHART

Here the reset index gives the new column starting from 0 in the frequency table and the total counts are calculated.

```
species_dist = df['species'].value_counts().reset_index()
species_dist
```

The column names are reassigned.

```
species_dist.columns=['species', 'count']
```

The species names to be displayed are defined in 'names' and percentage to be displayed are defined in'count'.

```
fig = px.pie(
    species_dist,
    names = 'species',
    values = 'count',
    title = 'Distribution of Species'
)
fig.show()
```

###INTERPRETATION
Pie chart represents the proportion of categorical data.The above graph shows the total species count.And it shows that they are equal counts for each species.
